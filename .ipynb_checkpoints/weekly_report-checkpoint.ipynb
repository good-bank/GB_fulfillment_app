```python
import pandas as pd
import seaborn as sns
import os
from pathlib import Path
import matplotlib.pyplot as plt

os.chdir(str(Path(os.getcwd()).parents[0]))


days = ["TUE", "WED", "THU", "FRI"]
week = 8
year = 2021
week_path = 'data/'+str(year)+'/CW'+str(week)+'/'
```

### Report for the week

#### Processed boxes


```python
df = pd.DataFrame()
for d in days:
    df = pd.concat([df, pd.read_csv(week_path+"extra_files/optimoroute_"+d+"_CW"+str(week)+"_all_columns.csv")])
df["TYPE"] = df["TYPE"].str.replace("\(1st box\)", "")
df.to_csv(week_path+"test.csv")

sns.countplot(x="processed_on", data=df)
plt.title("Total completed orders on week " +str(week));
plt.xlabel("")
plt.ylabel("Number of processed boxes")
```

#### New orders

```python
sns.countplot(x="processed_on", data=df.loc[df["type"]=="new",:])
plt.title("New orders on week " +str(week));
plt.xlabel("")
plt.ylabel("Number of processed boxes")
```

#### Total boxes counts

```python
df["TYPE"] = df["TYPE"].str.rstrip()
# total counts
majtypes = ["VEGAN", "VG", "OMNI"]
for tp in majtypes:
    print(tp +" boxes: "+str(df["TYPE"].str.contains(tp).sum()))
    
# counts including specials
sdf = df.groupby(by=["TYPE"]).count()
sdf["index"]
```

#### Boxes counts by day

```python
for d in days:
    print("DAY: "+d)
    tdf = df.loc[df["processed_on"]==d,:]
    # total counts
    majtypes = ["VEGAN", "VG", "OMNI"]
    for tp in majtypes:
        print(tp +" boxes: "+str(tdf["TYPE"].str.contains(tp).sum()))

    # counts including specials
    sdf = tdf.groupby(by=["TYPE"]).count()
    print(sdf["index"])
    
    print("")
```

### Geovizualizations

```python
import geopandas as gpd
plt.style.use('seaborn')
%matplotlib inline
```

```python
zip_shape = gpd.read_file('data/geodata/plz-gebiete.shp', dtype={'plz': str})
```

```python
#zip_shape.head()
loc_df = df.groupby(["ZIP"]).count().loc[:,["index"]]
loc_df["plz"] = loc_df.index
loc_df= loc_df.rename(columns={"index":"count"}).set_index("plz", drop=False)
zip_shape = zip_shape.set_index("plz", drop=False)
print(loc_df.columns)
print(zip_shape.columns)
loc_df = loc_df.join(zip_shape, how="left", lsuffix="_l")
loc_df
```

```python
plt.rcParams['figure.figsize'] = [16, 11]
fig, ax = plt.subplots()

zip_shape.plot(ax=ax, color='orange', alpha=0.8)
```

```python
fig, ax = plt.subplots()

loc_df.plot(
    ax=ax, 
    'count', 
    categorical=False, 
    legend=False, 
    cmap='autumn_r',
    alpha=0.8
)

    
ax.set(
    title='Germany: Number of Inhabitants per Postal Code', 
    aspect=1.3, 
    facecolor='lightblue'
);
```
